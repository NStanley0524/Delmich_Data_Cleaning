# Defined A Function For Effective Cleaning of The Columns

```python
# t as Datatable
# cname as Column name

# Remove Non-digits
def num_trim(t,cname):
    t[cname] = t[cname].str.replace(r'\D',"",regex= True)
    return t
    t = t.copy()

# Remove Digits
def obj_trim(t,cname):
    t[cname] = t[cname].str.replace(r'\d',"",regex= True)
    return t
    t = t.copy()

# Replace characters
def date_trim(t,cname):
    t[cname] = t[cname].str.replace("/","-")
    return t
    t = t.copy()

# Format date
def date_format(t,date):
    t[date] = pd.to_datetime(t[date],errors= "coerce",)
    t[date] = t[date].dt.strftime('%d-%m-%Y')
    return t
    t = t.copy()
```

## Checking Potential Outliers

### Age Column

```python
# Checking The Age Column for Potential Outliers

Q1_age = cleaned_df["Age"].quantile(0.25)
Q3_age = cleaned_df["Age"].quantile(0.75)
IQR = Q3_age - Q1_age

lower_bound_age = Q1_age - 1.5 * IQR
higher_bound_age = Q3_age + 1.5 * IQR

lower_bound_age,higher_bound_age
```

```python
age_outliers = cleaned_df[(cleaned_df["Age"] < lower_bound_age) | (cleaned_df["Age"] > higher_bound_age)]

age_outliers
```

```python
# Removing rows with the ages 4 or 78 to ensure data quality 

# Ages 4, 40 and 78 are not typical for a university-level course

cleaned_df = cleaned_df.drop(cleaned_df[(cleaned_df["Age"] == 4) | (cleaned_df["Age"] == 40) | (cleaned_df["Age"] == 78)].index) 
```

### Total Payments Column

```python
#Checking The Total Payments Column for Potential Outliers

Q1_payments = cleaned_df["Total_payments"].quantile(0.25)
Q3_payments = cleaned_df["Total_payments"].quantile(0.75)
IQR = Q3_payments - Q1_payments

lower_bound_payments = Q1_payments - 1.5 * IQR
higher_bound_payments = Q3_payments + 1.5 * IQR

lower_bound_payments,higher_bound_payments
```

```python
payment_outliers = cleaned_df[(cleaned_df["Total_payments"] < lower_bound_payments) | (cleaned_df["Total_payments"] > higher_bound_payments)]

payment_outliers.head()
```

```python
# Modifying the outliers with median payments

cleaned_df.loc[cleaned_df["Total_payments"] < lower_bound_payments, "Total_payments"] = median_payments
cleaned_df.loc[cleaned_df["Total_payments"] > higher_bound_payments, "Total_payments"] = median_payments
```
